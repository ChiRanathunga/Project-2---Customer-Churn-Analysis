```python
import pandas as pd

url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
(df.isnull()).sum()
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df=df.dropna()
```

```python
df['Churn'].value_counts()
```

```python
df.groupby('Contract')['Churn'].value_counts()
```

```python
df.groupby('Churn')['MonthlyCharges'].mean()
```

```python
df['InternetService'].value_counts()
```

```python
import matplotlib.pyplot as plt
df['Churn'].value_counts().plot(kind='bar')
plt.title("Customer Churn Count")
plt.show()
```

```python
df.groupby('Churn')['MonthlyCharges'].mean().plot(kind='bar',x = ['Churn'],y =['MounthlyCharges'])
plt.title("Monthly Charges Vs Churn")
plt.show()
```

# Insights

1. Most customers did not churn.
2. Monthly charges are higher for customers with churn = "Yes".
3. Most customers use Fiber Optic internet service.
4. Most customers have month-to-month contracts, both in the "Yes" and "No"  churn categories.

# Conclusion

This project analyzed customer churn behavior in a telecom company. The analysis revealed that most customers did not churn, while customers with higher monthly charges were more likely to churn. It was also observed that most customers used Fiber Optic internet service, and many customers in both churn categories had month-to-month contracts. This project demonstrates practical data analysis and business insight generation using Python and Pandas.
